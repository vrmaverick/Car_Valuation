# Encoding Categorical Data


```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

category = ["Vehicle_type","Model","Manufacturer"]#categorical labels
one_hot = OneHotEncoder() # instantiation
trans = ColumnTransformer([('one_hot',
                            one_hot,
                            category
                           )],remainder ='passthrough') # instantiation


X_new = trans.fit_transform(X) 
```

# Training a Regression Model

```python
# Building a regressor
from sklearn.ensemble import RandomForestRegressor 
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.linear_model import Ridge, Lasso
from sklearn.svm import SVR 
import matplotlib.pyplot as plt 

# Assuming you have your training data X_new and target variable Y

# Define models as dictionary entries for easier iteration
models = {
    'Random Forest': Pipeline([('regressor', RandomForestRegressor())]),
    'Ridge Regression': Pipeline([('regressor', Ridge())]),
    'Lasso Regression': Pipeline([('regressor', Lasso())]),
    'Support Vector Regression': Pipeline([('regressor', SVR())])
}

# Define a common hyperparameter grid for all models (you might need to adjust based on the models)
param_grid = {
    'Random Forest': {
        'regressor__n_estimators': [100, 200, 300], 
        'regressor__max_depth': [None, 10, 20]      
    },
    'Ridge Regression': {
        'regressor__alpha': [0.01, 0.1, 1.0]       
    },
    'Lasso Regression': {
        'regressor__alpha': [0.01, 0.1, 1.0]      
    },
    'Support Vector Regression': {
        'regressor__C': [0.1, 1.0, 10.0],           
        'regressor__kernel': ['linear', 'rbf']     
    }
}

# Perform GridSearchCV for each model
best_models = {}
for model_name, model in models.items():
    grid_search = GridSearchCV(model, param_grid[model_name], cv=5, n_jobs=-1)
    grid_search.fit(X_new,Y)
    # grid_search.fit(X_train, Y_train)
    best_models[model_name] = grid_search.best_estimator_

# Evaluate and compare the best models (choose your preferred evaluation metric)
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score  # Example metric

for model_name, model in best_models.items():
    y_pred = model.predict(X_new)  # Make predictions

    # Calculate MSE and MAE
    mse = mean_squared_error(Y, y_pred)
    mae = mean_absolute_error(Y, y_pred)

    # Accuracy for regression is not directly applicable, but you can consider R-squared
    r2 = r2_score(Y, y_pred)

    print(f"{model_name}:")
    print(f"\tMSE: {mse:.4f}")
    print(f"\tMAE: {mae:.4f}")
    print(f"\tR-squared: {r2:.4f}")
```
